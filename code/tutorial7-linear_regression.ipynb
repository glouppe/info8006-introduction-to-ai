```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Celsius_Farenheit(n_samp=20):
    W = np.array([9./5.]).reshape(1, 1)
    b = np.array([32.])
    x = np.random.uniform(low=-10, high=30, size=(n_samp, 1)) # T° in Celsius
    y = x @ W + b + np.random.randn(n_samp, 1) * 1.5 # Noised T° in Farhenheit
    return x, y
```

```python
x, y = Celsius_Farenheit(365)
```

```python
plt.figure(figsize=(20, 10))
plt.scatter(x[:, 0], y.reshape(-1))
```

```python
x_biased = np.concatenate((x, np.ones((x.shape[0], 1))), axis=1)
w_est = np.linalg.inv(x_biased.T @ x_biased)@x_biased.T @ y
w_est
```

```python
plt.figure(figsize=(20, 10))
plt.plot(x_biased[:, 0], x_biased @ w_est)
plt.scatter(x[:, 0], y.reshape(-1))
```

```python
residuals = x_biased @ w_est - y
plt.hist(residuals, bins=50)
plt.show()
residuals.mean(), residuals.std()
```
