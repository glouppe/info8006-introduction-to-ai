```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
plt.rcParams["figure.dpi"] = 160

def remove_frame():
    for spine in plt.gca().spines.values():
        spine.set_visible(False)
        
np.random.seed(43)
```

# Probability distributions

```python
from scipy.stats import uniform, norm
```

```python
xs = np.linspace(-5, 5, num=1000)
p = uniform(loc=0, scale=3)
plt.plot(xs, p.pdf(xs), c="r", label="${\cal U}(0,3)(x)$")
p = uniform(loc=-2, scale=4)
plt.plot(xs, p.pdf(xs), c="b", label="${\cal U}(-2,2)(x)$")
plt.grid()
plt.legend()
remove_frame()
plt.savefig("uniform.png")
```

```python
xs = np.linspace(-5, 5, num=1000)
p = norm(loc=0, scale=1)
plt.plot(xs, p.pdf(xs), c="r", label="${\cal N}(\mu=0,\sigma=1)(x)$")
p = norm(loc=1, scale=2)
plt.plot(xs, p.pdf(xs), c="b", label="${\cal N}(\mu=1,\sigma=2)(x)$")
plt.grid()
plt.legend()
remove_frame()
plt.savefig("normal.png")
```

```python
xs = np.linspace(-5, 5, num=1000)
p = norm(loc=0, scale=1)
plt.plot(xs, p.cdf(xs), c="r", label="$\Phi(\mu=0,\sigma=1)(x)$")
p = norm(loc=1, scale=2)
plt.plot(xs, p.cdf(xs), c="b", label="$\Phi(\mu=1,\sigma=2)(x)$")
plt.grid()
plt.legend()
remove_frame()
plt.savefig("probit.png")
```

```python
from scipy.stats import multivariate_normal
```

```python
from mpl_toolkits.mplot3d import Axes3D

mu_x = 0
variance_x = 1
mu_y = 0
variance_y = 5

x = np.linspace(-4,4,500)
y = np.linspace(-4,4,500)
X, Y = np.meshgrid(x,y)
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X; pos[:, :, 1] = Y
rv = multivariate_normal([mu_x, mu_y], [[variance_x, 0], [0, variance_y]])

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, rv.pdf(pos),cmap='Reds',linewidth=0)
ax.set_xlabel("$X_1$")
ax.set_ylabel("$X_2$")
ax.set_zlabel("$p(x_1,x_2)$")
remove_frame()
plt.savefig("mvn.png")
plt.show()
```
